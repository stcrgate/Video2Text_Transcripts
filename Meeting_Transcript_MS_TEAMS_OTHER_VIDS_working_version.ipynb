```python
# Step 1: Upload your video file
from google.colab import files
import os
import subprocess
import shlex

uploaded = files.upload()

# Ensure the file is uploaded
if not uploaded:
    raise ValueError("No file uploaded.")

# Get the name of the uploaded video file
input_video = list(uploaded.keys())[0]



# Step 2: Install necessary libraries
!pip install git+https://github.com/openai/whisper.git
!sudo apt update
!sudo apt install -y ffmpeg



# Step 3: Extract the audio from the video
output_audio = 'output_audio.mp3'
ffmpeg_command = f"ffmpeg -i {shlex.quote(input_video)} -q:a 0 -map a {output_audio}"

# Execute the FFmpeg command and capture the output
process = subprocess.run(ffmpeg_command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
stdout, stderr = process.stdout, process.stderr

# Print the FFmpeg command output for debugging
print(f"FFmpeg command: {ffmpeg_command}")
print(f"FFmpeg stdout: {stdout.decode('utf-8')}")
print(f"FFmpeg stderr: {stderr.decode('utf-8')}")

# Verify that the audio file has been created
if not os.path.exists(output_audio):
    raise RuntimeError(f"Failed to create audio file: {output_audio}")

# Step 4: Transcribe the audio using Whisper
import whisper

# Load the Whisper model
model = whisper.load_model("base")

# Path to your audio file
audio_file_path = output_audio

# Transcribe the audio file
result = model.transcribe(audio_file_path)

# Extract and save the transcript
transcript = result["text"]
with open("meeting_transcript.txt", "w") as f:
    f.write(transcript)

print("Transcription complete. The transcript has been saved to 'meeting_transcript.txt'.")

# Step 5: Download the transcript
files.download('meeting_transcript.txt')
```
